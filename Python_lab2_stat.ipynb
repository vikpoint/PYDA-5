### 6. Выдвинуть несколько гипотез по набору имеющихся переменных и проверить эти гипотезы с использованием t-test.

#### 6.1. Атмосферные двигатели по трассе расходают столько же сколько и турбированные

```python
atmo_hw = df[df['aspiration']=='std']['highway-mpg']
turbo_hw = df[df['aspiration']=='turbo']['highway-mpg']
```

```python
len(atmo_hw), len(turbo_hw)
```

```python
atmo_hw = atmo_hw.sample(35)
turbo_hw = turbo_hw.sample(35)
```

```python
statistics.variance(atmo_hw), statistics.variance(turbo_hw)
```

```python
sts.ttest_ind(atmo_hw, turbo_hw, equal_var=False)
```

**Вывод:** pvalue<0.05, гипотеза отвергается. Автомобили с атмосферными двигателями не расходают столько же сколько и турбированные.

#### 6.2. Автомобили с передним приводом стоят столько же сколько и с задним

```python
f_wd = df[df['drive-wheels']=='fwd']['price']
r_wd = df[df['drive-wheels']=='rwd']['price']
```

```python
len(f_wd), len(r_wd)
```

```python
f_wd = f_wd.sample(35)
r_wd = r_wd.sample(35)
```

```python
statistics.variance(f_wd), statistics.variance(r_wd)
```

```python
sts.ttest_ind(atmo_hw, turbo_hw, equal_var=False)
```

**Вывод:** pvalue<0.05, гипотеза отвергается. Автомобили с передним приводом не стоят столько же сколько и с задним.

#### 6.3. Автомобили с двумя дверьми расходуют топлива в городе столькоже сколько и автомобили с 4 дверьми

```python
two_doors = df[df['num-of-doors']=='two']['city-mpg']
four_doors = df[df['num-of-doors']=='four']['city-mpg']
```

```python
len(two_doors), len(four_doors)
```

```python
two_doors = two_doors.sample(35)
four_doors = four_doors.sample(35)
```

```python
statistics.variance(two_doors), statistics.variance(four_doors)
```

```python
sts.ttest_ind(two_doors, four_doors, equal_var=False)
```

**Вывод:** pvalue>0.05 гипотеза подтверждается. Автомобили с двумя дверьми расходуют топлива в городе столькоже сколько и автомобили с 4 дверьми.

#### 6.4. Автомобили с кузовом седан имееют такую же мощность, что и автомобили с остальными кузовами

```python
set(df['body-style'])
```
